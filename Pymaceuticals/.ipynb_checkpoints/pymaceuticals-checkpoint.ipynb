```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_df = pd.read_csv(mouse_drug_data_to_load)
drug_df = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
combined_df = pd.merge(drug_df, mouse_df, on='Mouse ID', how='left')

# Display the data table for preview
combined_df.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
tumor_data = combined_df.groupby(['Drug', 'Timepoint'])['Tumor Volume (mm3)']
mean_volume = tumor_data.mean()

# Convert to DataFrame
mean_volume_df = mean_volume.reset_index()

# Preview DataFrame
mean_volume_df.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
tumor_sem = tumor_data.sem()

# Convert to DataFrame
tumor_sem_df = tumor_sem.reset_index()

# Preview DataFrame
tumor_sem_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
# Pivot table required to create a scatter plot
pivot_mean_volume_df = mean_volume_df.pivot(index = 'Timepoint', columns = 'Drug', values = 'Tumor Volume (mm3)')
pivot_tumor_sem_df = tumor_sem_df.pivot(index = 'Timepoint', columns = 'Drug', values = 'Tumor Volume (mm3)')

# Preview that Reformatting worked
pivot_mean_volume_df.head()
```

```python
pivot_tumor_sem_df.head()
```

```python
# Generate the Plot (with Error Bars -- for the recommended studies)
plt.errorbar(pivot_mean_volume_df.index, pivot_mean_volume_df['Capomulin'], yerr=pivot_tumor_sem_df['Capomulin'], 
             marker='o', color='r', markersize=5, linestyle='--', linewidth=0.5)
plt.errorbar(pivot_mean_volume_df.index, pivot_mean_volume_df['Infubinol'], yerr=pivot_tumor_sem_df['Infubinol'], 
             marker='^', color='b', markersize=5, linestyle='--', linewidth=0.5)
plt.errorbar(pivot_mean_volume_df.index, pivot_mean_volume_df['Ketapril'], yerr=pivot_tumor_sem_df['Ketapril'], 
             marker='s', color='g', markersize=5, linestyle='--', linewidth=0.5)
plt.errorbar(pivot_mean_volume_df.index, pivot_mean_volume_df['Placebo'], yerr=pivot_tumor_sem_df['Placebo'], 
             marker='d', color='k', markersize=5, linestyle='--', linewidth=0.5)

plt.title('Tumor Response to Treatment')
plt.xlabel('Time (Days)')
plt.ylabel('Tumor Volume (mm3)')
plt.grid(axis='y')
plt.legend(loc='upper left')
# Save the Figure
plt.savefig('Images/treatment.png')
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
met_data = combined_df.groupby(['Drug', 'Timepoint'])['Metastatic Sites']
mean_met = met_data.mean()

# Convert to DataFrame
mean_met_df = pd.DataFrame(mean_met)

# Preview DataFrame
mean_met_df.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
met_sem = met_data.sem()

# Convert to DataFrame
met_sem_df = pd.DataFrame(met_sem)

# Preview DataFrame
met_sem_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
# Pivot table required to create a scatter plot
mean_met_df = mean_met_df.reset_index()
met_sem_df = met_sem_df.reset_index()
pivot_mean_met_df = mean_met_df.pivot(index = 'Timepoint', columns = 'Drug', values = 'Metastatic Sites')
pivot_met_sem_df = met_sem_df.pivot(index = 'Timepoint', columns = 'Drug', values = 'Metastatic Sites')
# Preview that Reformatting worked
# pivot_mean_met_df = pivot_mean_met_df.sort_values('Metastatic Sites', ascending=False)
pivot_mean_met_df.head()
```

```python
pivot_met_sem_df.head()
```

```python
# Generate the Plot (with Error Bars)
plt.errorbar(pivot_mean_met_df.index, pivot_mean_met_df['Capomulin'], yerr=pivot_met_sem_df['Capomulin'], 
             marker='o', color='r', markersize=5, linestyle='--', linewidth=0.5)
plt.errorbar(pivot_mean_met_df.index, pivot_mean_met_df['Infubinol'], yerr=pivot_met_sem_df['Infubinol'], 
             marker='^', color='b', markersize=5, linestyle='--', linewidth=0.5)
plt.errorbar(pivot_mean_met_df.index, pivot_mean_met_df['Ketapril'], yerr=pivot_met_sem_df['Ketapril'], 
             marker='s', color='g', markersize=5, linestyle='--', linewidth=0.5)
plt.errorbar(pivot_mean_met_df.index, pivot_mean_met_df['Placebo'], yerr=pivot_met_sem_df['Placebo'], 
             marker='d', color='k', markersize=5, linestyle='--', linewidth=0.5)

plt.title('Metastatic Spread During Treatment')
plt.xlabel('Time Duration (Days)')
plt.ylabel('Met. Sites')
plt.grid(axis='y')
plt.legend(loc='upper left')
# Save the Figure
plt.savefig('Images/spread.png')
# Show the Figure
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
mouse_data = combined_df.groupby(['Drug', 'Timepoint'])['Mouse ID']
mouse_count = mouse_data.nunique()

# Convert to DataFrame
mouse_count_df = mouse_count.reset_index()

# Rename column of Mouse ID to Mouse Count
mouse_count_df.rename(columns = {'Mouse ID':'Mouse Count'}, inplace=True)

# Preview DataFrame
mouse_count_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
pivot_mouse_df = mouse_count_df.pivot(index = 'Timepoint', columns = 'Drug', values = 'Mouse Count')

# Preview the Data Frame
pivot_mouse_df.head()
```

```python
# Generate the Plot (Accounting for percentages)
plt.errorbar(pivot_mouse_df.index, pivot_mouse_df['Capomulin']/mouse_count['Capomulin', 0]*100, marker='o', 
             color='r', markersize=5, linestyle='--', linewidth=0.5)
plt.errorbar(pivot_mouse_df.index, pivot_mouse_df['Infubinol']/mouse_count['Infubinol', 0]*100, marker='^', 
             color='b', markersize=5, linestyle='--', linewidth=0.5)
plt.errorbar(pivot_mouse_df.index, pivot_mouse_df['Ketapril']/mouse_count['Ketapril', 0]*100, marker='s', 
             color='g', markersize=5, linestyle='--', linewidth=0.5)
plt.errorbar(pivot_mouse_df.index, pivot_mouse_df['Placebo']/mouse_count['Placebo', 0]*100, marker='d', 
             color='k', markersize=5, linestyle='--', linewidth=0.5)

plt.title('Survival During Treatment')
plt.xlabel('Time (Days)')
plt.ylabel('Survival Rate (%)')
plt.grid(axis='y')
plt.legend(loc='bottom left')

# Save the Figure
plt.savefig('Images/survival.png')
# Show the Figure
plt.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
percent_change = ((pivot_mean_volume_df.iloc[-1] - pivot_mean_volume_df.iloc[0]) / pivot_mean_volume_df.iloc[0])*100

# Display the data to confirm
percent_change
```

```python
# Store all Relevant Percent Changes into a Tuple
drug_tuple = [('Capomulin', percent_change['Capomulin']), ('Infubinol', percent_change['Infubinol']), 
              ('Ketapril', percent_change['Ketapril']), ('Placebo', percent_change['Placebo'])]

x, y = zip(*drug_tuple)

# Splice the data between passing and failing drugs
passing = percent_change < 0

# Orient widths. Add labels, tick marks, etc. 
drug_bar = plt.bar(x, y, width=-1, align='edge', color=passing.map({True: 'g', False: 'r'}))
plt.title('Tumor Change Over 45 Day Treatment')
plt.ylabel('Survival Rate (%)')
plt.grid()
plt.ylim(-30, 70)

# Use functions to label the percentages of changes 
# (thanks to https://matplotlib.org/3.1.1/gallery/lines_bars_and_markers/barchart.html)
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        if height > 0:
            label_position = 2
        else:
            label_position = -8
        plt.text(rect.get_x() + rect.get_width() / 2, label_position,
                 '%d' % int(height)+'%', color='white',
                 ha='center', va='bottom')
        

# Call functions to implement the function calls
autolabel(drug_bar)

# Save the Figure
plt.savefig('Images/change.png')

# Show the Figure
drug_bar
```
